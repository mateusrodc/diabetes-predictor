```python
import sys
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sys.path.append(os.path.abspath('../src'))
from data_preprocessing import load_and_clean_data

df = load_and_clean_data('../data/diabetes.csv')
df.head()

# Informações gerais do dataset
print(df.info())

# Estatísticas descritivas
df.describe()

# Verificar valores nulos
print(df.isnull().sum())

# Verificar colunas com valores "0" que podem ser ausências disfarçadas
colunas_suspeitas = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in colunas_suspeitas:
    print(f'{col} tem {df[df[col] == 0].shape[0]} registros com valor zero')



# Distribuição da variável alvo
sns.countplot(data=df, x='Outcome')
plt.title('Distribuição de Pessoas com e sem Diabetes')
plt.show()

# Histograma para cada feature
df.hist(bins=20, figsize=(12, 10), color='#2f4f4f')
plt.suptitle('Distribuição das Variáveis', y=1.02)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlação entre Variáveis')
plt.show()
```
